## Cómo generar incrustaciones de texto con FastEmbed

### Importar modulos
En este ejemplo, utilizará el modelo de incrustación de texto predeterminado, BAAI/bge-small-en-v1.5.

```python
from fastembed import TextEmbedding
import numpy as np
```

### Agregar datos de muestra

Ahora, agregue dos documentos de muestra. Sus documentos deben estar en una lista y cada documento debe ser una cadena.

```python
documents = [
    "FastEmbed is lighter than Transformers & Sentence-Transformers.",
    "FastEmbed is supported by and maintained by Qdrant.",
]
```

Descargue e inicialice el modelo. Imprima un mensaje para verificar el proceso.

```python
embedding_model = TextEmbedding()
print("The model BAAI/bge-small-en-v1.5 is ready to use.")
```

```python
embedding_model_2 = TextEmbedding(model_name='jinaai/jina-embeddings-v2-small-en')
print("The model jinaai/jina-embeddings-v2-small-en is ready to use.")
```

### Incrustar datos

Generar incrustaciones para ambos documentos.

```python
embeddings_generator = embedding_model.embed(documents)
embeddings_list = list(embeddings_generator)
type(embeddings_generator)
```

```python
embeddings_generator_2 = embedding_model_2.embed(documents)
embeddings_list_2 = list(embeddings_generator_2)
type(embeddings_generator_2)
```
