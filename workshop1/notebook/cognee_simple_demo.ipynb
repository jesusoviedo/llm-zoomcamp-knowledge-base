Give the downloaded file location to cognee to save it and process its contents

```python
import cognee
# NOTE: As google colab is a free cloud service, the time it takes to cognify the alice in wonderland
#       book is around 10 minutes, compared to 2 minutes it takes to run locally
await cognee.add("file://alice_in_wonderland.txt")
await cognee.cognify()
```

Your data is ready to be queried:

```python
await cognee.search("List me influential characters in Alice in Wonderland.")
```
